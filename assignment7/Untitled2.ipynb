```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv('D:/DBDA_Official10/Machine_Learning/Machine_Learning/Assignment/Datasets/Salary_Train.csv')
df_test = pd.read_csv('D:/DBDA_Official10/Machine_Learning/Machine_Learning/Assignment/Datasets/Salary_Test.csv')
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df_train['Salary'] = encoder.fit_transform(df_train['Salary'])

df_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df_train['native'] = encoder.fit_transform(df_train['native'])
df_train['sex'] = encoder.fit_transform(df_train['sex'])
df_train['race'] = encoder.fit_transform(df_train['race'])
df_train['relationship'] = encoder.fit_transform(df_train['relationship'])
df_train['occupation'] = encoder.fit_transform(df_train['occupation'])
df_train['maritalstatus'] = encoder.fit_transform(df_train['maritalstatus'])
df_train['education'] = encoder.fit_transform(df_train['education'])
df_train['workclass'] = encoder.fit_transform(df_train['workclass'])
```

```python
x = df_train.drop(['Salary'],axis=1)
y = df_train['Salary']
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df_test['native'] = encoder.fit_transform(df_test['native'])
df_test['sex'] = encoder.fit_transform(df_test['sex'])
df_test['race'] = encoder.fit_transform(df_test['race'])
df_test['relationship'] = encoder.fit_transform(df_test['relationship'])
df_test['occupation'] = encoder.fit_transform(df_test['occupation'])
df_test['maritalstatus'] = encoder.fit_transform(df_test['maritalstatus'])
df_test['education'] = encoder.fit_transform(df_test['education'])
df_test['workclass'] = encoder.fit_transform(df_test['workclass'])
df_test['Salary'] = encoder.fit_transform(df_test['Salary'])
```

```python
x_test = df_test.drop(['Salary'] , axis = 1 )
y_test = df_test['Salary']
```

```python
def logistic():
    from sklearn.linear_model import LogisticRegressionCV
    model = LogisticRegressionCV()
    model.fit(x,y)
    return model

def naivebayes():
    from sklearn.naive_bayes import GaussianNB
    model = GaussianNB()
    model.fit(x,y)
    return model

def KNN():
    from sklearn.neighbors import KNeighborsClassifier
    model = KNeighborsClassifier()
    model.fit(x,y)
    return model

def svm():
    from sklearn.svm import SVC
    model = SVC()
    model.fit(x,y)
    return model

def decision():
    from sklearn.tree import DecisionTreeClassifier
    model = DecisionTreeClassifier()
    model.fit(x,y)
    return model

def random():
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier()
    model.fit(x,y)
    return model

def adaboost():
    from sklearn.ensemble import AdaBoostClassifier
    model = AdaBoostClassifier()
    model.fit(x,y)
    return model
```

```python
models = []

models.append({"name": "logistic" , "Model" : logistic()})
models.append({"name" : "naivebayes" , "Model" : naivebayes()})
models.append({ "name" : "KNN", "Model" : KNN()})
models.append({ "name" : "SVM" , "Model":svm()})
models.append({ "name" : "decision" ,"Model":decision()})
models.append({ "name" : "random" , "Model" : random()})
models.append({ "name" : "adaboost" , "Model" : adaboost()})
```

```python
from sklearn.metrics import accuracy_score , precision_score , recall_score , f1_score

for model in models:
    y_pred = model["Model"].predict(y_test)
    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred)
    recall = recall_score(y_test,y_pred)
    f1 = f1_score(y_test,y_pred)

    model["Accuracy"] = accuracy
    model["Precision"] = precision
    model["recall"] = recall
    model["f1"] = f1
```
