```python
Q1 = df['Absenteeism Time in Hours'].quantile(0.25)
Q3 = df['Absenteeism Time in Hours'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
```

```python
from sklearn.preprocessing import MinMaxScaler

# create an object of MinMaxScaler
minmax_scaler = MinMaxScaler()

# fit and transform the values
scaled_values = minmax_scaler.fit_transform(df.drop(['Absenteeism Time in Hours'],axis=1))

# convert the values to dataframe
df_scaled = pd.DataFrame(scaled_values, columns=['Reason for Absence','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets'])
df_new = pd.concat([df_scaled,df['Absenteeism Time in Hours']],axis=1)

df_new.head(3)
```
